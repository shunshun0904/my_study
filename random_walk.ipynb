```python
import numpy as np
import matplotlib.pyplot as plt

iteration = 1000

class RandomWalk():
    def __init__(self,iteration):
        self.iteration=iteration
        self.now_loc = 0
        self.history = [self.now_loc]
        self.direction = [1,-1]
        
    def progress(self):
        for i in range(self.iteration):
            self.now_loc = self.now_loc + np.random.choice(self.direction)
            self.history.append(self.now_loc)
      
    def plot(self):
        plt.plot(self.history)
        plt.show()
```

```python
test = RandomWalk(1000)
test.progress()
test.plot()
```
