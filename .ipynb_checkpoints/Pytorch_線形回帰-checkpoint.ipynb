## 勾配を計算してみよう

x = 1の時  

$y = x^2+2x$ という数式を考える。微分すると　$\frac{dy}{dx}= 2x+ 2$

```python
x = torch.tensor(1.0, requires_grad=True)
y = x**2+2*x
```

ニューラルネットワークを簡単に記述するためのライブラリである。
誤差逆伝播処理を実施した後に勾配計算するように内部の処理が組まれている。逆伝搬しないと（backward()しないと）勾配計算されないようになっているので、勾配計算（微分計算）する際は基本的にbackward()→grad　する

```python
y.backward()
x.grad
```

## ボストンハウス

ここまでの復讐として、ボストンの家の値段を回帰予測するモデル（線形回帰モデル）を実装してみる。
　線形回帰なので先ほど、同様、層の数は1層。

```python
from torch.utils.data import TensorDataset, DataLoader 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# データの読み込み
boston = load_boston()
x_data = np.array(boston.data)
y_data = np.array(boston.target)

# トレインとテストに分割
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3)

# それzれ標準化を行う
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

# それぞれtorchテンソルに変換する
x_train = torch.Tensor(x_train)
x_test = torch.Tensor(x_test)
y_train = torch.Tensor(y_train)
y_test = torch.Tensor(y_test)
 
#データとラベルをセットにしたDatasetを作成
ds_train = TensorDataset(x_train, y_train)
ds_test = TensorDataset(x_test, y_test)
 
#データセットのミニバッチサイズを指定した、Dataloaderを作成
# Chainerのiterators.SerialIteratorと似ている
loader_train = DataLoader(ds_train, batch_size=32, shuffle=True)
loader_test = DataLoader(ds_test, batch_size=32, shuffle=True)
```

## 以下でニューラルネットを定義する
今は特徴量が13個なので、(13,1)の重みをもつ層を定義する

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(13,1)
        
    def forward(self, x):
        x = self.fc1(x)
        return x
```

- modelをインスタンス化
- 損失関数の定義
- 最適化手法の定義

```python
model = Model()
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.003)
```

## 学習してみる

面倒くさいので、検証データを作成せず（パラメーターチューニングせず）に一発で学習し、その学習モデルで予測する

```python
num_epoch = 300 
```

```python
loss_list = []
for epoch in range(num_epoch):  
    #データ全てのトータルロス
    running_loss = 0.0 
    for i, data in enumerate(loader_train):
        #入力データ・ラベルに分割
        # get the inputs
        inputs, labels = data
        # Variableに変形
        inputs, labels = Variable(inputs), Variable(labels)
        # optimizerの初期化
        optimizer.zero_grad()

        #一連の流れ
        outputs = model(inputs)

        out = outputs.view(-1, 1)
        labels = labels.view(-1, 1)
        loss = criterion(outputs, labels) 
        loss.backward()
        optimizer.step()

        # ロスの表示
        running_loss += loss.item()
        
        # 20ミニバッチごとにロスの表示
        if i % 10 == 9:    
            print('[%d, %5d] loss: %.3f' %
                 (epoch + 1, i + 1, running_loss / 10))
            loss_list.append(running_loss)
            running_loss = 0.0
            
```

```python
plt.plot(loss_list)
plt.show()
```

## 推定してみる

```python
pred = model(x_test)
pred
```

## 推定結果に対する考察

```python
pred2 = pred.reshape(-1).tolist()
plt.hist(y_test, bins = 40 , color = "r")
plt.hist(pred2, bins = 40, color = "b")
plt.show()
```

線形モデルなので、データの分布は正規分布になる。正規分布から外れているサンプル（上図の端）では、当てはまりの悪いモデルになっている。
